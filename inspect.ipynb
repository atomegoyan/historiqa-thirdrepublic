```python
import pandas as pd
import json
import os
data_dir = os.getcwd() + '/dataset/'


# read json line file
def load_data(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))
    return data

df_corpus = pd.DataFrame(load_data(os.path.join(data_dir, 'corpus_dataset.jsonl')))
```
